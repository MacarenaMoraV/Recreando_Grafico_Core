# Visualización para ciencia de datos / Recreando gráfico (Core)

* 1. Carga del Dataset: Usa la función sns.load_dataset(‘tips’) para cargar el dataset de ejemplo ‘tips’ de Seaborn.
* 2. Creación de la Figura y el Eje: Utiliza plt.subplots para crear una figura de tamaño adecuado.
* 3. Gráfico de Dispersión: Completa la llamada a sns.scatterplot añadiendo las columnas correctas para los parámetros hue, style y s, y asegúrate de asignar el gráfico al eje correcto (ax).
* 4. Cuadrícula: Asegúrate de que la cuadrícula esté configurada correctamente utilizando los parámetros which, linestyle, linewidth y color.
* 5. Anotaciones: Completa las coordenadas xy y xytext para ubicar la anotación correctamente.
* 6. Flechas: Completa las coordenadas iniciales y la dirección de la flecha con los valores correctos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset
df = sns.load_dataset('tips')

# Crear figura y eje
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de dispersión
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    style='smoker',
    s=100,
    ax=ax
)

# Título y etiquetas
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)

# Leyenda con fondo del color del gráfico
ax.legend(loc='upper right', fontsize='large', shadow=False, title='Detalles')
leg = ax.get_legend()
leg.get_frame().set_alpha(0.3)
leg.get_frame().set_linewidth(0)
leg.get_frame().set_facecolor('white')



# Cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Punto con mayor tip
mayor_tip = df.loc[df['tip'].idxmax()]
x_anot = mayor_tip['total_bill']
y_anot = mayor_tip['tip']

# Dibujar el punto manualmente para resaltarlo
ax.scatter(x_anot, y_anot, color='black', s=200, marker='*', zorder=5, label='Mayor tip')

# Añadir anotación
ax.annotate( "Mayor tip en la cena", xy=(x_anot, y_anot), xytext=(x_anot + 2, y_anot + 2),
    arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir flecha
ax.arrow(30, 4.9,10, 3.1, head_width=0.6, head_length=0.3, fc='green',ec='green')

# Mostrar el gráfico
plt.show()
```
